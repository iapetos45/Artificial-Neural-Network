```python
import numpy as np

def Or(x1, x2):
    x = np.array([x1, x2]) # 입력
    w = np.array([0.5, 0.5]) # 가중치
    b = -0.2 # bias (편향)
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1
```

```python
Or(0, 0)
```

```python
Or(1, 0)
```

```python
Or(0, 1)
```

```python
Or(1, 1)
```
