```python
import numpy as np

def And(x1, x2):
    x = np.array([x1, x2]) # 입력
    w = np.array([0.5, 0.5]) # 가중치
    b = -0.7 # bias (편향)
    tmp = np.sum(w*x + b)
    if tmp <= 0:
        return 0
    else:
        return 1

def Nand(x1, x2):
    x = np.array([x1, x2]) # 입력
    w = np.array([-0.5, -0.5]) # 가중치
    b = 0.7 # bias (편향)
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

def Or(x1, x2):
    x = np.array([x1, x2]) # 입력
    w = np.array([0.5, 0.5]) # 가중치
    b = -0.2 # bias (편향)
    tmp = np.sum(w*x) + b
    if tmp <= 0:
        return 0
    else:
        return 1

def Xor(x1, x2):
    s1 = Nand(x1, x2)
    s2 = Or(x1, x2)
    y = And(s1, s2)
    return y
```

```python
Xor(0, 0)
```

```python
Xor(1, 0)
```

```python
Xor(0, 1)
```

```python
Xor(1, 1)
```
